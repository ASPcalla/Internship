```python
g = sns.pairplot(data)
g.fig.suptitle('FacetGrid plot', fontsize = 20)
g.fig.subplots_adjust(top= 0.9);
```

```python
# Lets create a list of attributes 
gdp_happiness = ['Happiness Score','Economy (GDP per Capita)']

# lets create a dataframe that only contains these attributes
gdp_corr = data[gdp_happiness]

# Lets Find the correlation
gdp_corr.corr()
```

```python
sns.regplot(data = gdp_corr, x = 'Happiness Score', y = 'Economy (GDP per Capita)').set_title("Correlation-Happiness score vs Economy/GDP")
```
