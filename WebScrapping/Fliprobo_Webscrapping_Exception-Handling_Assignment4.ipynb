Q1. 
Scrape the details of most viewed videos on YouTube from Wikipedia. Url = 
https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:

Rank, Name, Artist, Upload Date , Views

```python
driver = webdriver.Chrome()

# Opening the flipkart.com pag Son the automated chome browser
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
time.sleep(2)
#maximize the window
driver.maximize_window()
```

```python
Rank=[]
Name=[]
Artist=[]
Upload_date=[]
Views=[]
```

```python
#scraping data
try:
    rank_ele=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
    for i in rank_ele[0:30]:
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append('NA')
except StaleElementReferenceException:
    Rank.append('NA')
    
# Extracting Name Via X path
try:
    name_ele=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
    for i in name_ele:
        Name.append(i.text)
except NoSuchElementExceptionhElementException:
    Name.append('NA')
except StaleElementReferenceException:
    Name.append('NA')
    
# Extracting Artist Name Via Xpath
try:
    artist_ele=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")
    for i in artist_ele:
        Artist.append(i.text)
except NoSuchElementExceptionhElementException:
    Artist.append('NA')
except StaleElementReferenceException:
    Artist.append('NA')

# Extracting Upload date Via Xpath
try:
    upload_date_ele=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")
    for i in upload_date_ele:
        Upload_date.append(i.text)
except NoSuchElementExceptionhElementException:
    Upload_date.append('NA')
except StaleElementReferenceException:
    Upload_date.append('NA') 

# Extracting Views via Xpath
try:
    views_ele=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")
    for i in views_ele:
        Views.append(i.text)
except NoSuchElementExceptionhElementException:
    Views.append('NA')
except StaleElementReferenceException:
    Views.append('NA')
    

```

```python
print(len(Rank),len(Name),len(Artist),len(Views),len(Upload_date))
```

```python
max_length = max(len(Rank), len(Name), len(Artist), len(Views), len(Upload_date))

# Pad shorter lists with default values (e.g., None or empty string) to make them the same length
Rank         += [None] * (max_length - len(Rank))
Name         += [None] * (max_length - len(Name))
Artist       += [None] * (max_length - len(Artist))
Views        += [None] * (max_length - len(Views))
Upload_date  += [None] * (max_length - len(Upload_date))
```

```python
# printing Dataframe for scrap data
wiki_video=pd.DataFrame({'Rank':Rank,'Video Name':Name,'Uploader':Artist,'Views (in Billons)':Views,'Upload Date':Upload_date})
print('\033[1m'+'Most Viewed Video on YouTube from Wikipedia :'+'\033[0m')
wiki_video
```

```python
driver = webdriver.Chrome()

# Opening the flipkart.com pag Son the automated chome browser
driver.get("https://www.imdb.com/list/ls095964455/")
time.sleep(3)
#maximize the window
driver.maximize_window()
```

```python
#Extracting Name:
name =[i.text for i in driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]/a')]

# Extracting Year Span
year = [i.text for i in driver.find_elements(By.XPATH,'//span[@class="lister-item-year text-muted unbold"]')]

#Extracting Genre 
genre = [i.text for i in driver.find_elements(By.XPATH,'//p[@class="text-muted text-small"]/span[5]')]

#Extracting Run time
time = [i.text for i in driver.find_elements(By.XPATH,'//p[@class="text-muted text-small"]/span[3]')]

#Extracting Rating
rating = [i.text for i in driver.find_elements(By.XPATH,'//div[@class="ipl-rating-widget"]/div/span[2]')]

#Extracting Votes
voting = [i.text for i in driver.find_elements(By.XPATH,'//p[@class="text-muted text-small"][3]/span[2]')]

driver.close()
print(len(name), len(year), len(genre), len(time), len(rating),len(voting))

#Creating DataFrame
most_watched_tv_series=pd.DataFrame({
    "Name": name,
    "Year span":year,
    "Genre":genre,
    "Run time":time,
    "Ratings":rating,
    "Votes": voting
})
most_watched_tv_series
```
